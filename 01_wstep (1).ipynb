## ‚öôÔ∏è Dwa g≈Ç√≥wne podej≈õcia w AI

Z biegiem lat w ramach AI wykszta≈Çci≈Çy siƒô dwa du≈ºe podej≈õcia:

1. **Machine Learning (uczenie maszynowe)**
2. **Rule-based AI (sztuczna inteligencja oparta na regu≈Çach)**

Oba dƒÖ≈ºƒÖ do inteligentnych zachowa≈Ñ, ale r√≥≈ºniƒÖ siƒô tym, **w jaki spos√≥b ‚Äûwiedza‚Äù trafia do systemu**.

---

### ü§ñ 1. Machine Learning (ML)

To podej≈õcie, w kt√≥rym **system uczy siƒô z danych**.
Zamiast programowaƒá zasady dzia≈Çania rƒôcznie, **dostarczamy przyk≈Çady**, a algorytmy uczƒÖ siƒô wzorc√≥w i podejmujƒÖ decyzje samodzielnie.

üí¨ *Przyk≈Çad:*
Pokazujemy tysiƒÖce zdjƒôƒá kot√≥w i ps√≥w ‚Üí model sam nauczy siƒô rozpoznawaƒá, gdzie kot, a gdzie pies.

ML to dzisiaj **najwa≈ºniejsza i najszybciej rozwijajƒÖca siƒô czƒô≈õƒá AI**, obejmujƒÖca takie obszary jak sieci neuronowe, SVM, drzewa decyzyjne czy deep learning.

---

### ‚öñÔ∏è 2. Rule-based AI (AI oparte na regu≈Çach)

To starsze, klasyczne podej≈õcie do sztucznej inteligencji.
Tutaj **cz≈Çowiek rƒôcznie zapisuje wiedzƒô w postaci regu≈Ç logicznych** typu:

> JE≈öLI pacjent ma gorƒÖczkƒô ORAZ b√≥l gard≈Ça,
> TO podejrzewaj infekcjƒô.

System nie uczy siƒô z danych ‚Äî **dzia≈Ça dok≈Çadnie wed≈Çug zapisanych zasad**.
Wiedza jest jawna, interpretowalna i w pe≈Çni kontrolowana przez cz≈Çowieka.

---

## üîπ Przyk≈Çady system√≥w rule-based AI

### üß© Systemy ekspertowe (Expert Systems)

To programy, kt√≥re na≈õladujƒÖ decyzje ludzkich ekspert√≥w z danej dziedziny (np. medycyny, prawa, finans√≥w).
Dzia≈ÇajƒÖ na zasadzie **bazy regu≈Ç** i **mechanizmu wnioskowania**, kt√≥ry wybiera odpowiednie regu≈Çy na podstawie danych wej≈õciowych.

üí¨ *Przyk≈Çad:*
System medyczny diagnozuje choroby na podstawie objaw√≥w, u≈ºywajƒÖc tysiƒôcy regu≈Ç zapisanych przez lekarzy.

---

### üí¨ Rule-based NLP (Natural Language Processing)

To przetwarzanie jƒôzyka oparte na **gramatyce i regu≈Çach jƒôzykowych**, nie na uczeniu.
System rozpoznaje s≈Çowa, formy gramatyczne i sens, korzystajƒÖc z zestawu zasad opracowanych rƒôcznie przez lingwist√≥w.

üí¨ *Przyk≈Çad:*

 
## üß† 1. Co to jest Machine Learning (ML)

**Uczenie maszynowe (ML)** to dziedzina AI, w kt√≥rej **system uczy siƒô samodzielnie z danych**,
zamiast byƒá zaprogramowany na sztywno.

> Model ML dostaje dane wej≈õciowe i przyk≈Çady poprawnych wynik√≥w,
> i uczy siƒô dopasowywaƒá swojƒÖ ‚Äûfunkcjƒô‚Äù tak, by przewidywaƒá poprawne odpowiedzi.

üí¨ Innymi s≈Çowy:

> ‚ÄûNie uczymy komputera *co* robiƒá, tylko *dajemy mu dane*, ≈ºeby sam to odkry≈Ç.‚Äù

## ‚öôÔ∏è 2. G≈Ç√≥wne typy uczenia maszynowego

Uczenie maszynowe dzielimy na trzy g≈Ç√≥wne kategorie:

```
Machine Learning
‚îÇ
‚îú‚îÄ‚îÄ Supervised Learning (uczenie nadzorowane)
```

## üîπ 3. Supervised Learning (uczenie nadzorowane)

To **najczƒôstszy typ ML**.

Dajemy modelowi:

* dane wej≈õciowe (X),
* poprawne odpowiedzi (y) ‚Üí tzw. **etykiety (labels)**.

Model uczy siƒô odwzorowania:
$[
X \Rightarrow y
]$
czyli: ‚Äûdla takich danych wej≈õciowych ‚Äì taka odpowied≈∫‚Äù.

### üìò Przyk≈Çady:

* rozpoznawanie obraz√≥w (X = zdjƒôcie, y = etykieta ‚Äûkot‚Äù lub ‚Äûpies‚Äù)
* przewidywanie ceny domu (X = cechy domu, y = cena)
* analiza sentymentu (X = tekst, y = ‚Äûpozytywne‚Äù lub ‚Äûnegatywne‚Äù)

### üîß Typowe algorytmy:

* Regresja liniowa / logistyczna
* Drzewa decyzyjne, Random Forest
* SVM (Support Vector Machines)
* k-NN (k-Nearest Neighbors)
* Sieci neuronowe (MLP, CNN, RNN)

---

## üîπ 4. Unsupervised Learning (uczenie nienadzorowane)

Tutaj **nie ma etykiet (y)**.
Dajemy tylko dane (X), a model sam szuka **struktur, wzorc√≥w lub grup**.

> Celem jest odkrycie ukrytych zale≈ºno≈õci ‚Äî bez wiedzy o poprawnych odpowiedziach.

### üìò Przyk≈Çady:

* grupowanie klient√≥w wg zachowa≈Ñ zakupowych
* wykrywanie anomalii (np. oszustwa finansowe)
* redukcja wymiarowo≈õci (np. PCA)

### üîß Typowe algorytmy:

* K-Means (klasteryzacja)
* PCA (analiza g≈Ç√≥wnych sk≈Çadowych)
* Autoenkodery (w deep learningu)

---

## üîπ 5. Reinforcement Learning (uczenie przez wzmacnianie)

Tutaj **nie ma gotowych danych z etykietami**,
tylko **≈õrodowisko**, w kt√≥rym **agent** uczy siƒô poprzez do≈õwiadczenie**:

* wykonuje akcjƒô,
* dostaje **nagrodƒô lub karƒô**,
* i uczy siƒô, kt√≥re decyzje prowadzƒÖ do najlepszych efekt√≥w.

üí¨ To uczenie przez pr√≥bƒô i b≈ÇƒÖd ‚Äî jak dziecko uczƒÖce siƒô chodziƒá.

### üìò Przyk≈Çady:

* gry (np. **AlphaGo**, **Chess AI**, **Atari**)
* robotyka (sterowanie ruchem)
* optymalizacja (np. systemy rekomendacji, logistyka)

### üîß Typowe algorytmy:

* Q-Learning, Deep Q-Network (DQN)
* Policy Gradient
* PPO (Proximal Policy Optimization ‚Äì u≈ºywany np. w ChatGPT)

---

## ü§ñ 6. Gdzie pasujƒÖ LLM-y (Large Language Models)

**LLM (np. GPT-4, Claude, Gemini, LLaMA)** to modele jƒôzykowe, kt√≥re uczƒÖ siƒô rozumienia i generowania jƒôzyka naturalnego.
To **hybryda r√≥≈ºnych podej≈õƒá ML**, ale g≈Ç√≥wny trzon to **Supervised / Self-supervised Learning** z elementami **Reinforcement Learning (RLHF)**.

üìò Etapy uczenia LLM:

| Etap             | Typ uczenia            | Co siƒô dzieje                                                                           |
| ---------------- | ---------------------- | --------------------------------------------------------------------------------------- |
| **Pre-training** | Self-supervised        | model uczy siƒô przewidywaƒá kolejne s≈Çowo w tek≈õcie (np. po miliardach zda≈Ñ z Internetu) |
| **Fine-tuning**  | Supervised             | dostrajanie do zada≈Ñ typu Q&A, streszczanie, t≈Çumaczenie                                |
| **RLHF**         | Reinforcement Learning | uczenie z ocen ludzi (nagroda za lepsze odpowiedzi)                                     |

üí¨ W skr√≥cie:

> LLM to przyk≈Çad **self-supervised learning** (specjalny przypadek supervised),
> z elementami **reinforcement learning** w ko≈Ñcowym etapie (RLHF).

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_classification

# --- Dane z make_blobs ---
X_blobs, y_blobs = make_blobs(
    n_samples=300,
    centers=3,              # trzy wyra≈∫ne klastry
    n_features=2,
    random_state=42,
    cluster_std=1.0
)

# --- Dane z make_classification ---
X_class, y_class = make_classification(
    n_samples=300,
    n_features=2,           # dwie cechy (do wizualizacji)
    n_informative=2,        # obie cechy zawierajƒÖ informacjƒô
    n_redundant=0,          # brak cech nadmiarowych
    n_repeated=0,
    n_classes=3,            # trzy klasy
    n_clusters_per_class=1, 
    class_sep=1.2,          # separacja miƒôdzy klasami
    random_state=42
)

# --- Wykresy ---
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# make_blobs
axes[0].scatter(X_blobs[:, 0], X_blobs[:, 1], c=y_blobs, cmap=plt.cm.coolwarm, s=40, edgecolor='k')
axes[0].set_title("Dane z make_blobs() ‚Äî proste, czyste klastry")
axes[0].set_xlabel("Cecha 1")
axes[0].set_ylabel("Cecha 2")

# make_classification
axes[1].scatter(X_class[:, 0], X_class[:, 1], c=y_class, cmap=plt.cm.coolwarm, s=40, edgecolor='k')
axes[1].set_title("Dane z make_classification() ‚Äî bardziej realistyczne")
axes[1].set_xlabel("Cecha 1")
axes[1].set_ylabel("Cecha 2")

plt.tight_layout()
plt.show()
````
